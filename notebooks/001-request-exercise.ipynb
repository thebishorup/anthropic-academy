### Goal:

I want system to prompt to write something in the input field and then I want to print the content of the message object. Also i want to answer the question and print the content of the response. While doing this the conversation should be maintained.

```python
# Initialize API Client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
def add_user_message(messages, content):
    messages.append({
        "role": "user",
        "content": content
    })
    
def add_assistant_message(messages, content):
    messages.append({
        "role": "assistant",
        "content": content
    })
    
def chat(messages):
    # Multi Conversation
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages = messages
    )
    return message.content[0].text
```

```python
messages = []

# Prompt the user for input
while True:
    user_input = input("You: ")
    print("You: ", user_input)
    
    # Add user message to the conversation
    add_user_message(messages, user_input)
    
    response = chat(messages)
    print("Assistant: ", response)
    
    # Add assistant message to the conversation
    add_assistant_message(messages, response)
```
