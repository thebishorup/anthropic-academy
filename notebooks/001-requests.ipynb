```python
# Create an API Client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
# Make a Request
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages = [
        {
            "role": "user",
            "content": "What is the meaning of life?"
        }
    ]
)
```

```python
message.content[0].text
```

```python
def add_user_message(messages, content):
    messages.append({
        "role": "user",
        "content": content
    })
    
def add_assistant_message(messages, content):
    messages.append({
        "role": "assistant",
        "content": content
    })
    
def chat(messages):
    # Multi Conversation
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages = messages
    )
    return message.content[0].text
```

```python
# Make start of conversation
messages = []

# Add user message
add_user_message(messages, "What is the meaning of life?")

# Pass the list of message to the chat function
response = chat(messages)

# Take the response and add it to the messages list
add_assistant_message(messages, response)

# Add the user's follow up question
add_user_message(messages, "What is the meaning of life according to Aristotle?")

# Make a follow up request with the updated messages list
response = chat(messages)

response
```
