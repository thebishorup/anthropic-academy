````python
# Create an API Client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
````

````python
def add_user_message(messages, content):
    messages.append({
        "role": "user",
        "content": content
    })
    
def add_assistant_message(messages, content):
    messages.append({
        "role": "assistant",
        "content": content
    })
    
def chat(messages, system = None, temperature = 1.0, stop_sequences = None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }
    
    if stop_sequences:
        params["stop_sequences"] = stop_sequences
    
    if system:
        params["system"] = system
    
    # Multi Conversation
    message = client.messages.create(**params)
    
    return message.content[0].text
````

````python
messages = []

system = '''
You are a senior software engineer with expertise in python language.
You are mentoring the a junior software engineer about DS and algorithms.
When you anwser the question explain the solution with code example step by step.
You explanations should be simple and easy to understand.
'''

add_user_message(messages, "Write a puython function to reverse a string")

response = chat(messages, system=system)
response
````
